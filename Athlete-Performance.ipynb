Reading the CSV File

```python
import pandas as pd
import numpy as np
df = pd.read_csv("../../data/BU Athletics Data Set v.4.csv", low_memory=False)
pd.set_option("display.max_columns",None)
df.head()
```

Cleaning the Data

```python
#if the "US Region" is not filled then fill it as "International"
df['US Region'] = df['US Region'].fillna('International')
df.head()
```

```python
#drop all columns if the entire column only has 0 values
df = df.loc[:, (df != 0).any(axis=0)]
df[df['US Region'].isin(["Northeast", 'West', 'South', 'Midwest',"Northeast"])]['English Language'].unique()
```

```python
#populate the language for US natives as English Native
df['English Language'].fillna("English Native",inplace=True)
df.head()
```

```python
#drop the rows which have missing HSGPA values
df.dropna(subset=['HSGPA'], inplace=True)
df.head()
```

```python
df['Entering Term'].unique()

#clean duplicate values for 'Entering Term'
df['Entering Term'].replace("Fall2016","Fall 2016",inplace=True)
df['Entering Term'].replace("Fall2017","Fall 2017",inplace=True)
df['Entering Term'].replace("Fall2018","Fall 2018",inplace=True)
df['Entering Term'].replace("Fall2019","Fall 2019",inplace=True)
df['Entering Term'].replace("Fall2020","Fall 2020",inplace=True)
df['Entering Term'].replace("Fall2021","Fall 2021",inplace=True)
df['Entering Term'].replace("Sprg2021","Spring 2021",inplace=True)
df['Entering Term'].replace("Sprg2022","Spring 2022",inplace=True)
df['Entering Term'].unique()
```

```python
#clean FCLASS and retension class
df['RetentionClass'].unique()
df['RetentionClass'].replace("Freshman","Freshmen",inplace=True)
```

```python
df[(df['RetentionClass'].str.lower()!=df['FCLASS'].str.lower())]
```

```python
df['First_Major'] = df['FIRSTCOLLEGE'] + df['FIRSTMAJOR'].astype('str')
df['LAST_Major'] = df['LASTCOLLEGE'] + df['LASTMAJOR'].astype('str')
ff = df[['First_Major','LAST_Major']]
ff.head()
```

'FCLASS' and 'RetensionClass' differ only for graduate students.

1. What is the range of accepted SAT/ACT scores and highschool GPA for student athletes?

```python
import matplotlib.pyplot as plt

# Add axis labels and a legend
bins = pd.cut(df['SATCFORAI'], range(400, 1601, 100))

# Group the data by the bins and count the number of values in each group
grouped_data = df.groupby(bins)['SATCFORAI'].count()

# Plot a bar graph for each range of SATCFORAI values
grouped_data.plot(kind='bar',color="orange", edgecolor="magenta")

# Set the plot title and axis labels
plt.title('Distribution of SAT Scores')
plt.xlabel('SAT Score Range')
plt.ylabel('Number of Students')

# Show the plot
plt.show()

satcforai_min = df['SATCFORAI'].min()
satcforai_max = df['SATCFORAI'].max()

print('Minimum value in SATCFORAI column:', satcforai_min)
print('Maximum value in SATCFORAI column:', satcforai_max)
print("Hence the range for accepted SAT scores are from", satcforai_min, "to", satcforai_max)
```

Range of accepted highschool GPA's

```python
bins = [2, 2.5, 3, 3.5, 4, 4.3]
# Plot a histogram of the High school gpa column
df['HSGPA'].plot.hist(bins=bins, color="orange", edgecolor="magenta", rwidth=0.5)

# Set the plot title and axis labels
plt.title('Distribution of High School GPA Scores')
plt.xlabel('High School GPA Range')
plt.ylabel('Number of Students')
# Show the plot
plt.show()

hsgpa_min = df['HSGPA'].min()
hsgpa_max = df['HSGPA'].max()

print('Minimum GPA recorded in given data:', hsgpa_min)
print('Maximum GPA recorded in given data:', hsgpa_max)
print("Hence the range for accepted High School GPA scores are from", hsgpa_min, "to", hsgpa_max)
```

```python
counts = df['HSGPA'].value_counts().sort_index()

#dropping less significant values
labels=["<=3.0"]
less_than_3=0
gr_than_4=0
total=823 #total students
for key,value in counts.items():
    per=round(100*value/total,1)
    if key<3.1:
        less_than_3+=per
        counts=counts.drop(key)
    elif key>=4.0:
        gr_than_4+=per
        counts=counts.drop(key)
    else:
        labels.append(key)
counts=pd.concat([pd.Series(data=less_than_3,index=["<=3.0"]),counts])
counts.append(pd.Series(data=gr_than_4,index=[">=4.0"]))
labels.append(">=4.0")
```

```python

# create a pie chart
counts.plot.pie(autopct='%1.1f%%', startangle=90, textprops={'fontsize': 8},figsize=(7,7),pctdistance=0.7,labeldistance=1.1)
plt.axis('equal')
plt.legend(labels=labels, loc="lower right")
plt.title('Distribution of High School Grades')
plt.show()
```

1.2 Does English being the primary language of the country impact the students' performance?

```python

# Calculate the impact of language on GPA of students
english_percentage = df[df['English Language']=='English Native']['Last Cumulative BU GPA'].mean()

# Calculate the percentage of non-international students
non_english_percentage = df[df['English Language']!='English Native']['Last Cumulative BU GPA'].mean()

# Create a bar chart showing the percentage of international students and non-international students
plt.bar(['Non-English', 'English'], [english_percentage, non_english_percentage], color="orange", edgecolor="magenta")
plt.title('Average CGPA of English speakers vs Non-English speakers')
plt.xlabel('Language')
plt.ylabel('Average CGPA')
plt.show()
print("The average CGPA for native English speakers is:" + str(round(english_percentage,1)))
print("The average CGPA for non native English speakers is:" + str(round(non_english_percentage,1)))
```

```python
# Subset the data by enrollment year
enrollment_years = df['Entering Term'].unique()

# Calculate the average CGPA for native English speakers and non-native English speakers for each enrollment year subset
english_percentage_by_year = []
non_english_percentage_by_year = []

for year in enrollment_years:
    english_percentage_by_year.append(df[(df['Entering Term']==year) & (df['English Language']=='English Native')]['Last Cumulative BU GPA'].mean())
    non_english_percentage_by_year.append(df[(df['Entering Term']==year) & (df['English Language']!='English Native')]['Last Cumulative BU GPA'].mean())

ind = np.arange(len(enrollment_years))
width = 0.3

fig, ax = plt.subplots(figsize=(10,6))
rects1 = ax.bar(ind - width/2, english_percentage_by_year, width, label='English Native', color="orange", edgecolor="magenta")
rects2 = ax.bar(ind + width/2, non_english_percentage_by_year, width, label='Non-English', color="blue", edgecolor="magenta")

ax.set_title('Average CGPA of English speakers vs Non-English speakers by Enrollment Year')
ax.set_xlabel('Enrollment Year')
ax.set_ylabel('Average CGPA')
ax.set_xticks(ind)
ax.set_xticklabels(enrollment_years)
ax.legend()

plt.show()
```

```python
english_speakers = df[df['English Language'] == 'English Native']
non_english_speakers = df[df['English Language'] != 'English Native']

english_gpa = english_speakers.loc[:, 'Cumulative GPA-1st Sem':'Cumulative GPA-22nd Sem'].mean()
non_english_gpa = non_english_speakers.loc[:, 'Cumulative GPA-1st Sem':'Cumulative GPA-22nd Sem'].mean()

x_labels = ["sem " + str(i)  for i in range(1, 23)]

plt.plot(x_labels, english_gpa, label='English Native', color="orange")
plt.plot(x_labels, non_english_gpa, label='Non-English', color="blue")

plt.xticks(rotation=90)
plt.xlabel("Semester")
plt.ylabel("Average GPA")
plt.legend()
plt.show()
```

Since the average CGPA for native english speakers and non-native students is almost the same, we can say that Language does not play a strong role in the CGPAs of students.

3) How do these students with high SAT/ACT scores or high GPA perform academically at BU compared to student athletes with low SAT/ACT scores or low GPA?

```python
df.head()
```

```python
# Ignoring students who don't have a SAT/ACT score
submitted_scores = df[df['SATCFORAI'].notnull()]
no_scores = df[df['SATCFORAI'].isnull() == True]
no_scores['SATCFORAI'].fillna(900,inplace=True)
# Assuming that the top 10% of scores are considered high SAT/ACT scores
top_sat_score = submitted_scores['SATCFORAI'].quantile(0.9)
top_sat_scorers = submitted_scores[submitted_scores['SATCFORAI'] >= top_sat_score]
other_sat_scorers = submitted_scores[submitted_scores['SATCFORAI'] < top_sat_score]
# Seperating scores between top scorers and other scorers
top_sat_scorers_CGPA = top_sat_scorers['Cumulative GPA-22nd Sem']
top_sat_scorers_SATC = top_sat_scorers['SATCFORAI']
other_sat_scorers_CGPA = other_sat_scorers['Cumulative GPA-22nd Sem']
other_sat_scorers_SATC = other_sat_scorers['SATCFORAI']
no_sat_scorers_CGPA = no_scores['Cumulative GPA-22nd Sem']
no_sat_scorers_SATC = no_scores['SATCFORAI']
# Plotting a scatterplot
fig, ax = plt.subplots()
ax.scatter(top_sat_scorers_SATC, top_sat_scorers_CGPA, label='Top SAT Scorers')
ax.scatter(other_sat_scorers_SATC, other_sat_scorers_CGPA, label='Other SAT Scorers')
ax.scatter(no_sat_scorers_SATC, no_sat_scorers_CGPA, label='Did not provide a SAT score',cmap='gray')
ax.set_xlabel('SAT Score')
ax.set_ylabel('Cumulative GPA')
ax.set_title('Comparison of SAT Scores and Cumulative GPAs')
ax.legend()
plt.show()
print("Average cgpa for students who did not have a SATCFORAI",no_sat_scorers_CGPA.mean())
print("Average cgpa for students with high SATCFORAI",top_sat_scorers_CGPA.mean())
print("Average cgpa for other students",other_sat_scorers_CGPA.mean())
```

```python
# Assuming that the top 10% of GPA's are considered high GPA scores
top_GPA_score = submitted_scores['HSGPA'].quantile(0.9)
top_GPA_scorers = submitted_scores[submitted_scores['HSGPA'] >= top_GPA_score]
other_GPA_scorers = submitted_scores[submitted_scores['HSGPA'] < top_GPA_score]
# Seperating scores between top scorers and other scorers
top_GPA_scorers_CGPA = top_GPA_scorers['Cumulative GPA-22nd Sem']
top_GPA_scorers_GPA = top_GPA_scorers['HSGPA']
other_GPA_scorers_CGPA = other_GPA_scorers['Cumulative GPA-22nd Sem']
other_GPA_scorers_GPA = other_GPA_scorers['HSGPA']
no_sat_scorers_CGPA = no_scores['Cumulative GPA-22nd Sem']
no_sat_scorers_GPA = no_scores['HSGPA']
# Plotting a scatterplot
fig, ax = plt.subplots()
ax.scatter(top_GPA_scorers_GPA, top_GPA_scorers_CGPA, label='Top HSGPA Scorers')
ax.scatter(other_GPA_scorers_GPA, other_GPA_scorers_CGPA, label='Other HSGPA Scorers')
ax.scatter(no_sat_scorers_GPA, no_sat_scorers_CGPA, label='HSGPA for student with no SAT score',alpha=0.4)
ax.set_xlabel('High School GPA')
ax.set_ylabel('Cumulative GPA')
ax.set_title('Comparison of High School GPA Scores and Cumulative GPAs')
ax.legend()
plt.show()
print("Average cgpa for students who did not have a SATCFORAI",no_sat_scorers_CGPA.mean())
print("Average cgpa for students with the highest HSGPA",top_GPA_scorers_CGPA.mean())
print("Average cgpa for other students",other_GPA_scorers_CGPA.mean())
```

4. Does the average student athlete improve their GPA throughout their college year?

```python
# Create a list of demographic groups to plot
demographic_groups = list(df['IPEDS Race Description'].unique())

# Create a plot for each demographic group
for demographic_group in demographic_groups:
    # Create a new DataFrame that contains only the data for the chosen demographic group
    group_df = df.loc[df['IPEDS Race Description'] == demographic_group]
    
    # Calculate the average GPA for each semester
    average_semester_gpa = []
    for x in group_df.loc[:, 'Cumulative GPA-1st Sem':'Cumulative GPA-22nd Sem']:
        average_semester_gpa.append(group_df[x].mean())

    # Plot the results
    x_labels = ["sem " + str(i) for i in range(1, 23)]
    plt.plot(x_labels, average_semester_gpa, label=demographic_group)

# Add plot labels and legend
plt.xticks(rotation=90)
plt.xlabel("Semester")
plt.ylabel("Average GPA")
plt.title("Average GPA by Demographic Group")
plt.legend(title="Demographic Group", loc="best")

# Show the plot
plt.show()
```

The average GPA for students does increase with each semester, but gets constant after a while.

7. What are the most common majors for student athletes? Do student athletes tend to switch majors? What percentage of student athletes stay in the same major?

```python
def college_major(series):
    First_Major = series["First_Major"]
    LAST_Major = series["LAST_Major"]
    change = 0 if First_Major == LAST_Major else 1
    return change

ff["Change"] = ff.apply(college_major,axis=1)
percentage = ((sum(ff['Change'])/len(ff['Change']))*100,((len(ff['Change'])-sum(ff['Change']))/len(ff['Change']))*100)
plt.figure()
plt.title('Student athletes who changed major vs. who not changed')
plt.bar(('Change', 'NotChange'), percentage)
plt.ylabel('Percentage')
plt.xlabel('ChangeOrNot')
print(sum(ff['Change'])/len(ff['Change']))
```

```python
plt.figure(figsize = (10,5))
index = count.index.tolist()
percentage = count.tolist()
percentage.pop()
index.pop()
for i in range(len(percentage)):
    percentage[i] = (int(percentage[i]) / sum_count)*100
index = np.array(index)
percentage = np.array(percentage)
index = index[:10]
percentage = percentage[:10]
plt.bar(index,percentage)
plt.title('Number of student athletes in different major(Top 10)')
plt.ylabel('Percentage')
plt.xlabel('Major')
print(percentage)
```
