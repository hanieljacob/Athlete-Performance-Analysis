```python
import pandas as pd
import numpy as np
df = pd.read_csv("BU Athletics Data Set v.4.csv", low_memory=False)
pd.set_option("display.max_columns",None)
df.head()
```

Cleaning the Data

```python
#if the "US Region" is not filled then fill it as "International"
df['US Region'] = df['US Region'].fillna('International')
df.head()
```

```python
#drop all columns if the entire column only has 0 values
df = df.loc[:, (df != 0).any(axis=0)]
df[df['US Region'].isin(["Northeast", 'West', 'South', 'Midwest',"Northeast"])]['English Language'].unique()
```

```python
#populate the language for US natives as English Native
df['English Language'].fillna("English Native",inplace=True)
df.head()
```

```python
#drop the rows which have missing HSGPA values
df.dropna(subset=['HSGPA'], inplace=True)
df.head()
```

```python
df['Entering Term'].unique()

#clean duplicate values for 'Entering Term'
df['Entering Term'].replace("Fall2016","Fall 2016",inplace=True)
df['Entering Term'].replace("Fall2017","Fall 2017",inplace=True)
df['Entering Term'].replace("Fall2018","Fall 2018",inplace=True)
df['Entering Term'].replace("Fall2019","Fall 2019",inplace=True)
df['Entering Term'].replace("Fall2020","Fall 2020",inplace=True)
df['Entering Term'].replace("Fall2021","Fall 2021",inplace=True)
df['Entering Term'].replace("Sprg2021","Spring 2021",inplace=True)
df['Entering Term'].replace("Sprg2022","Spring 2022",inplace=True)
df['Entering Term'].unique()
```

```python
#clean FCLASS and retension class
df['RetentionClass'].unique()
df['RetentionClass'].replace("Freshman","Freshmen",inplace=True)
```

```python
counts = df['HSGPA'].value_counts().sort_index()

#dropping less significant values
labels=["<=3.0"]
less_than_3=0
gr_than_4=0
total=823 #total students
for key,value in counts.items():
    per=round(100*value/total,1)
    if key<3.1:
        less_than_3+=per
        counts=counts.drop(key)
    elif key>=4.0:
        gr_than_4+=per
        counts=counts.drop(key)
    else:
        labels.append(key)
counts=pd.concat([pd.Series(data=less_than_3,index=["<=3.0"]),counts])
counts.append(pd.Series(data=gr_than_4,index=[">=4.0"]))
labels.append(">=4.0")
```

7. What are the most common majors for student athletes? Do student athletes tend to switch majors? What percentage of student athletes stay in the same major?

```python
def college_major(series):
    First_Major = series["First_Major"]
    LAST_Major = series["LAST_Major"]
    change = 0 if First_Major == LAST_Major else 1
    return change

ff["Change"] = ff.apply(college_major,axis=1)
percentage = ((sum(ff['Change'])/len(ff['Change']))*100,((len(ff['Change'])-sum(ff['Change']))/len(ff['Change']))*100)
plt.figure()
plt.title('Student athletes who changed major vs. who not changed')
plt.bar(('Change', 'NotChange'), percentage)
plt.ylabel('Percentage')
plt.xlabel('ChangeOrNot')
print(sum(ff['Change'])/len(ff['Change']))
```

```python
def college_major(series):
    First_Major = series["First_Major"]
    LAST_Major = series["LAST_Major"]
    change = 0 if First_Major == LAST_Major else 1
    return change

ff['Entering Term'] = df['Entering Term']

# Calculate percentage of students who changed vs not changed major for each entering term
term_groups = ff.groupby('Entering Term')
change_percentages = []
for term, group in term_groups:
    total_count = len(group)
    change_count = sum(group['Change'])
    change_percent = (change_count / total_count) * 100
    not_change_percent = 100 - change_percent
    change_percentages.append((change_percent, not_change_percent))

# Plot side by side bar chart for each entering term
bar_width = 0.35
fig, ax = plt.subplots()
ax.bar(np.arange(len(term_groups))-bar_width/2, [p[0] for p in change_percentages], bar_width, label='Changed')
ax.bar(np.arange(len(term_groups))+bar_width/2, [p[1] for p in change_percentages], bar_width, label='Not Changed')
ax.set_xticks(range(len(term_groups)))
ax.set_xticklabels([term for term, group in term_groups])
ax.set_title('Student athletes who changed major vs. who not changed by entering term')
ax.set_xlabel('Entering Term')
ax.set_ylabel('Percentage')
plt.xticks(rotation=90)
ax.legend()
plt.show()
```

```python
import scipy.stats as stats

# create two separate dataframes for students who did and did not change their major
no_change = df[ff['Change'] == 0]['Last Cumulative BU GPA']
change = df[ff['Change'] == 1]['Last Cumulative BU GPA']

# perform a two-sample t-test to compare the means of the two groups
t, p = stats.ttest_ind(no_change, change, equal_var=False)

# print the results
print("t-value:", t)
print("p-value:", p)
```

Based on the t-value of 1.95 and the p-value of 0.05, we can say that there is a marginally significant difference between the average college GPA of student-athletes who changed their major and those who did not. However, since the p-value is greater than the commonly used significance level of 0.05, we cannot reject the null hypothesis that there is no significant difference between the two groups. In other words, we cannot say with high confidence that changing one's major has a significant effect on their college GPA.
